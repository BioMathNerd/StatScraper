```python
import requests
```

```python
data_url = 'https://www.ncaa.com/stats/football/fbs/current/individual/469'
```

```python
response = requests.get(data_url)
```

```python
response.status_code
```

```python
len(response.text)
```

```python
page_contents = response.text
```

```python
page_contents[:1000]
```

```python
with open('webpage.html', 'w') as f:
    f.write(page_contents)
```

```python
from bs4 import BeautifulSoup
```

```python
doc = BeautifulSoup(page_contents, 'html.parser')
```

```python
doc
```

```python
td_tags = doc.find_all('td')
```

```python
th_tags = doc.find_all('th')
```

```python
len(th_tags)
```

```python
th_tags[:9]
```

```python
heading_info = []
for heading in th_tags:
    heading_info.append(heading.text)
print(heading_info)
```
